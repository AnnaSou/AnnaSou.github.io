In order to work with the data, we need to change columns manes for more convenient ones and merge data for 2018th year with 2016th year column 'Region'.
```python
happy1 = pandas.merge(happy2018,happy2016.loc[:,['Country', 'Region']], how = 'left', on = 'Country',)
happy1.head()
```

Before we change columns' names, let us explain their meanings.
It was found that happiness can be explained by the following six factor:
GDP  per capita is self explonatory
Social support, measured by having someone to count on in times of troubles
Healthy life expectancy - healthy years of life expectancy
Perceptions of corruption is meauserd by a percieved absence of corruption in goverment and business
Freedom to make life choices
Generosity as measured by recent donations

Taken together these six variables explained up to 3/4 of the variation in national annual average ladder score among countries using data for the years 2005 to 2016.

```python
happy2 = happy1.rename(columns={'Happiness score': 'Happiness_score', 'Dystopia (1.92) + residual': 'Residual','Explained by: GDP per capita': 'GDP','Explained by: Social support': 'Social_support', 'Explained by: Healthy life expectancy': 'Health','Explained by: Freedom to make life choices':'Freedom','Explained by: Generosity':'Generosity','Explained by: Perceptions of corruption':'Corruption'})
happy2.head()
```

One more step is to add data for population. It would be used for plotting purpouses. First, we scrap data from a website, then merge it with our data frame for 2018th year.

```python
data = pd.read_csv("gap.tsv", sep='\t')
data = data[data['year']==2007]
data = data.rename(columns={'country': 'Country'})
happy3 = pandas.merge(happy2,data.loc[:,['Country', 'pop']],how='left',on='Country')
happy3.head()
```

Checking missing values, and there are lots of them. So, we need to find another source in the web and get missing values from it.

```python
print("Are there any missing values for column 'Population'?: {}".format(len(happy3[happy3['pop'].isnull()])))
```

```python
link = "http://www.worldometers.info/world-population/population-by-country/"
f = requests.get(link)
text = f.text
soup = bs(f.text, "html.parser")
prettyHTML=soup.prettify() 
t = soup.find('table',class_="table table-striped")
tt = soup.find_all("table","table table-sm table-striped")
table = pandas.read_html(prettyHTML)
df=table[0].rename(index=str, columns={"#":"rank","Country (or dependency)": "Country","Population  (2018)":"pop"})
dff= df.loc[:,['Country','pop']]
```

After we merged dataframes, we still have some rows missing. Since there are not so many, we again input them manually.

```python
A = happy3[happy3['pop'].isnull()]
AB = pandas.merge(A,dff,how='left',on='Country')
AB= AB.drop('pop_x',axis=1)
AB[AB['pop_y'].isnull()]
```

```python
AB.loc[3,'pop_y'] = 23571000
AB.loc[5,'pop_y'] = 1369000
AB.loc[12,'pop_y'] = 200587
AB.loc[21,'pop_y'] = 7392000
AB.loc[17,'pop_y'] = 1831000
AB.loc[24,'pop_y'] = 2074000
AB.loc[27,'pop_y'] = 4685000
AB.loc[30,'pop_y'] = 5261000
AB.loc[33,'pop_y'] = 81339988
AB.loc[28,'pop_y'] = 24294750
AB = AB.rename(columns={'pop_y':'pop'})
```

Check for missing values:

```python
AB[AB['pop'].isnull()]
```

```python
res = pandas.merge(happy3,AB.loc[:,['Country','pop']],how='left',on='Country')
res2 = res
res2['pop_x'].update(res2['pop_y'])
res2 = res2.drop('pop_y',axis=1)
res2 = res2.rename(columns={'pop_x':'Population'})
```
